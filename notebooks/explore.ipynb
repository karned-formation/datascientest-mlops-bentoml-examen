```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df = pd.read_csv('../data/raw/admission.csv')
df
```

```python
df.drop(columns=['Serial No.'], inplace=True)
df
```

```python
df.describe()
```

```python
df.info()
```

```python
colonnes = ['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA',
       'Research']
```

```python
scaler_minmax = MinMaxScaler()
df[colonnes] = scaler_minmax.fit_transform(df[colonnes])

scaler_std = StandardScaler()
df[colonnes] = scaler_std.fit_transform(df[colonnes])
```
